```python
roi = image[40:150,50:200]
cv2_imshow(roi)
```

```python
resized = cv2.resize(image,(200,150))
cv2_imshow(image)
cv2_imshow(resized)
```

```python
output = image.copy()
cv2.rectangle(output,(75,70),(150,100),(40,100,255),2)
cv2_imshow(output)
```

```python
img1 = np.flipud(image)
cv2_imshow(img1)
```

```python
b,g,r = cv2.split(image)
cv2_imshow(r)
cv2_imshow(g)
cv2_imshow(b)
```

```python
img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
```

```python
import random

gaussian_noise = np.zeros((img.shape[0],img.shape[1]),dtype=np.uint8)

cv2_imshow(gaussian_noise)
cv2.randn(gaussian_noise,128,20)
cv2_imshow(gaussian_noise)
cv2.imwrite("img.jpg",gaussian_noise)
```

```python
gaussian_noise = (gaussian_noise * 0.5).astype(np.uint8)
noisy1 = cv2.add(img,gaussian_noise)



cv2_imshow(noisy1)
```

```python
blurred1 = cv2.medianBlur(noisy1,3)
st1 = np.hstack((noisy1,blurred1))
cv2_imshow(st1)
cv2.imwrite("Median filter_Gaussian noise.jpg",blurred1)

#blurred2 = cv2.medianBlur(noisy2,3)
#st2 = np.hstack(noisy2,blurred2)
#cv2_imshow(st2)
#cv2.imwrite("Median filter - uniform noise",blurred2)
```

```python
processed_img1 = cv2.blur(noisy1,(5,5))
st1 = np.hstack((noisy1,processed_img1))
cv2_imshow(st1)
```

```python
pi = cv2.bilateralFilter(noisy1,5,75,75)

cv2_imshow(pi)
```

```python
imggp = cv2.GaussianBlur(noisy1,(5,5),1)
st1 = np.hstack((noisy1,processed_img1,imggp))
cv2_imshow(st1)
```

```python
import matplotlib.pyplot as plt
edge = cv2.Canny(img,75,75)
cv2_imshow(edge)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(edge,cmap = 'gray')
plt.title('Edge Image'),plt.xticks([]),plt.yticks([])
plt.show()
```
